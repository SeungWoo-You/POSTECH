```python
import numpy as np
import matplotlib.pyplot as plt
from gtda.mapper import Projection, CubicalCover, make_mapper_pipeline
from gtda.mapper import plot_static_mapper_graph, plot_interactive_mapper_graph
from sklearn.cluster import DBSCAN
import gudhi as gd
from sklearn.preprocessing import MinMaxScaler
```

```python
rips_complex = gd.RipsComplex(points=[[1, 0], [np.sqrt(2), np.sqrt(2)], [0, 1], [-np.sqrt(2), np.sqrt(
    2)], [-1, 0], [-np.sqrt(2), -np.sqrt(2)], [0, -1], [np.sqrt(2), -np.sqrt(2)]], sparse=0.4)
simplex_tree = rips_complex.create_simplex_tree(max_dimension=2)
fig, axes = plt.subplots(ncols=2, nrows=1, figsize=(15, 5))
gd.plot_persistence_barcode(simplex_tree.persistence(), axes=axes[0])
gd.plot_persistence_diagram(simplex_tree.persistence(), axes=axes[1])
```

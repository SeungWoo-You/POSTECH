```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
F1 = 1
F2 = 1
N = 1000
res = [0] * 10
first = [0] * 10
first[1] += 2
for _ in range(3, N + 1):
    F = F1 + F2
    try:
        i = int(str(F)[1])
        res[i] += 1
    except:
        pass
    i = int(str(F)[0])
    first[i] += 1
    F1 = F2
    F2 = F
w = 0.4
plt.bar(np.linspace(0 - w / 2, 9 - w / 2, 10), first, width=w, alpha=0.8)
plt.bar(np.linspace(0 + w / 2, 9 + w / 2, 10), res, width=w, alpha=0.8)
Y = [np.log10(1 + 1 / x) * 1000 for x in range(1, 10)]
plt.plot(range(1, 10), Y, 'ro')
plt.xticks(range(10))
plt.legend(['Benford', 'first', 'second'])
plt.show()
```

```python
def Is_Prime(x):
    if x <= 3:
        if x <= 1:
            return False
        return True
    if not np.mod(x, 2) or not np.mod(x, 3):
        return False
    for k in range(5, int(x**0.5)+1, 6):
        if not np.mod(x, k) or not np.mod(x, k + 2):
            return False
        return True
```

```python
M = 100000
M = int(M)
N = 0
k = 0
num = np.zeros(9)
while N < M:
    k += 1
    if Is_Prime(k):
        N += 1
        # print(N, k)
        tmp = str(k)
        x = int(tmp[0])
        num[x - 1] += 1
        if N > 1 and float(np.log10(N)).is_integer():
            plt.plot(range(1, 10), num / sum(num), '-o', label=f'N = {N}')
Y = [np.log10(1 + 1 / x) for x in range(1, 10)]
plt.bar(range(1, 10), Y, color='r', alpha=0.2, label='Benford')
plt.xticks(range(1, 10))
plt.legend()
plt.show()
```

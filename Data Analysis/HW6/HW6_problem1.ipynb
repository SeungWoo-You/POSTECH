```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
city_data = np.loadtxt('pop.txt', dtype=str)
city_data = np.char.replace(city_data, ',', '').astype(np.float64)
c = len(city_data)
```

```python

plt.plot(np.log10(range(1, c + 1)), np.log10(city_data), 'ro', markersize=3)
plt.xlabel('Log Rank')
plt.ylabel('population')
plt.legend(['City population'])
plt.show()
print(np.polyfit(np.log10(range(1, c + 1)), np.log10(city_data), deg=1))
```

```python
C = 0
s, t = np.polyfit(np.log10(range(1 + C, 5 + 1 + C)),
                  np.log10(city_data[:5]), deg=1)
print(s, t)
m, M = min(np.log10(range(1 + C, c + 1 + C))
           ), max(np.log10(range(1 + C, c + 1 + C)))
Y = [s * x + t for x in np.linspace(m, M, 100)]
plt.plot(np.log10(range(1, c + 1)), np.log10(city_data), 'ro', markersize=3)
plt.xlabel('Log Rank')
plt.ylabel('population')
plt.plot(np.linspace(m, M, 100), Y)
p, q = np.polyfit(np.log10(range(1, c + 1)), np.log10(city_data), deg=1)
A = [p * x + q for x in np.linspace(m, M, 100)]
plt.plot(np.linspace(m, M, 100), A, 'darkgray', linestyle='--', linewidth=2)
plt.legend(['City population', 'y = -1.05x + 6.94', 'y = -1.30x + 7.38'])
```

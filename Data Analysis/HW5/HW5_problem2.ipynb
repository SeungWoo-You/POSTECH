```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
mu = (np.e-(np.e**(-1)))/2  # exact mean value
M = 100
dM = 100
Error = []  # error about mean; exact mean = 0
MM = []  # sample number
for i in range(100):
    M += 100
    X = [[-1 + (j - 1) * (2 / (M - 1)) for j in range(1, M + 1)]]
    Y = np.exp(X)
    mu_L = np.sum(Y) / M
    MM.append(M)
    Error.append(abs(mu_L - mu))
plt.plot(np.log10(MM), np.log10(Error))
slope, inter = np.polyfit(np.log10(MM), np.log10(Error), 1)
print(slope, inter)
```
